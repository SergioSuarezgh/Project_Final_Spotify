```python
def direccion2(url,dia):

    
    driver=webdriver.Chrome(PATH)
    driver.get(url)
    
    time.sleep(2)

    boton_login=driver.find_element(By.XPATH,'//*[@id="__next"]/div/div/main/div[2]/div/header/div/div[2]/a[3]/div[1]').click()
    
    time.sleep(2)
    
    input_mail=driver.find_element(By.XPATH,'//*[@id="login-username"]')
    
    input_mail.send_keys(user_spoty)  # escribe texto

    input_clave=driver.find_element(By.XPATH,'//*[@id="login-password"]')
    
    
    input_clave.send_keys(pass_spoty)
    
    boton_acceso=driver.find_element(By.XPATH,'//*[@id="login-button"]/div[1]/p').click()
    
    time.sleep(2)
    
    dias = pd.date_range(start='2017-01-01', end='2017-01-02') 
    dfs=[]
    
    for d in dias:
        driver.get('https://charts.spotify.com/charts/view/regional-global-daily/'+d.strftime("%Y-%m-%d"))
        
        time.sleep(2)
        
        
        tabla=driver.find_element(By.XPATH,'//*[@id="__next"]/div/div/main/div[2]/div[3]/div/table/tbody')
        
        
        time.sleep(2)
        

                
        for i in range(1,3):
            
            linea=tabla.find_element(By.XPATH,f'//*[@id="__next"]/div/div/main/div[2]/div[3]/div/table/tbody/tr[{i}]/td[3]')
            
            
            
            boton=linea.find_element(By.XPATH,f'//*[@id="__next"]/div/div/main/div[2]/div[3]/div/table/tbody/tr[{i}]/td[8]/span/button')
            
    
            tabla_int=driver.find_element(By.XPATH,f'//*[@id="__next"]/div/div/main/div[2]/div[3]/div/table/tbody/tr[{i}]')
            
            filas=tabla.find_elements(By.TAG_NAME, 'tr')

            
            data=[]

            for f in filas:

                elementos=f.find_elements(By.TAG_NAME, 'td') 

                tmp=[]

                for e in elementos:

                    tmp.append(e.text)
                
                
                data.append(tmp)
        df=pd.DataFrame(data, columns=['index', 'novedadx','titulox','pos_Actual','pos_Anterior','semanas_lista', 'streams','unnamed1','unnamed2'])
        
        df['dia']=d.strftime("%Y-%m-%d")
        
        df.novedadx=df.novedadx.apply(lambda x: x.split('\n'))
        df.titulox=df.titulox.apply(lambda x: x.split('\n'))
        
        df['posicion']=None
        df['novedad']=None
        df['titulo']=None
        df['autor']=None
        
        for i in range(len(df.novedadx)):
            df['posicion'][i]=df.novedadx[i][0]
            df['novedad'][i]=df.novedadx[i][1]
            df.titulo[i]=df.titulox[i][0]
            df.autor[i]=df.titulox[i][1]
            
        df = df.reindex(columns=['index','dia','posicion','novedad','titulo','autor','pos_Actual','pos_Anterior','semanas_lista',
                                'streams','unnamed1','unnamed2'])
        dfs.append(df)

    return dfs
```
