```python
def direccion(url,semana,ciudad):

    
    
    
    driver.get(f'https://charts.spotify.com/charts/view/citytoptrack-{ciudad}-weekly/{semana}')
    
    
    
    time.sleep(3)
    
    
    tabla=driver.find_element(By.XPATH,'//*[@id="__next"]/div/div/main/div[2]/div[3]/div/table/tbody')
    
    
    time.sleep(3)
    
                            
    filas=tabla.find_elements(By.TAG_NAME, 'tr')

        
    data=[]

    for f in filas:

        elementos=f.find_elements(By.TAG_NAME, 'td') 

        tmp=[]

        for e in elementos:

            tmp.append(e.text)
            
            
        data.append(tmp)
    df=pd.DataFrame(data)
    
    df.rename(columns={1:'position',2:'autor_cancion',3:'peak_rank',4:'previous_rank',5: 'days_on_chart'},inplace=True)
    df.position=df.position.apply(lambda x: x.split('\n')[0])
    df['artist_names']=df.autor_cancion
    df['track_name']=df.autor_cancion
    df['fecha']=semana

    df.autor_cancion=df.autor_cancion.apply(lambda x: x.split('\n'))

    for i in range(len(df)):
        df['artist_names'][i]=df.autor_cancion[i][1]
        df['track_name'][i]=df.autor_cancion[i][0]
    
    df.drop(columns='autor_cancion', inplace=True)

    df=df.reindex(columns=[ 'fecha','artist_names','track_name','position','previous_rank','peak_rank','days_on_chart'])


    
    


    return df
```
